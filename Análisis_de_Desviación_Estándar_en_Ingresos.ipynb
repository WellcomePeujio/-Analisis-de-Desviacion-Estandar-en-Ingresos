# Análisis de Desviación Estándar en Ingresos

## Problema

Una empresa ha registrado los ingresos mensuales durante el último año (en miles de dólares):

| Mes        | Ingresos ($) |
|------------|--------------|
| Enero      | 50           |
| Febrero    | 45           |
| Marzo      | 55           |
| Abril      | 60           |
| Mayo       | 65           |
| Junio      | 48           |
| Julio      | 52           |
| Agosto     | 70           |
| Septiembre | 68           |
| Octubre    | 55           |
| Noviembre  | 62           |
| Diciembre  | 58           |

## Preguntas:

1. **Calcula la media y la desviación estándar de los ingresos mensuales.**
2. **¿Qué mes se desvía más de la media y qué significa esto para la empresa?**

# Resolución Matemática: Análisis de Desviación Estándar en Ingresos

## Datos del problema:

Los ingresos mensuales (en miles de dólares) durante el último año son:

- Enero: $50
- Febrero: $48
- Marzo: $52
- Abril: $47
- Mayo: $55
- Junio: $49
- Julio: $53
- Agosto: $51
- Septiembre: $46
- Octubre: $54
- Noviembre: $50
- Diciembre: $56

Representamos los ingresos como un conjunto de datos \$( X \$):

$$
X = \{50, 48, 52, 47, 55, 49, 53, 51, 46, 54, 50, 56\}
$$

### Pregunta 1: Cálculo de la media y la desviación estándar

### Paso 1: Cálculo de la media

La fórmula para la media \$( \mu \$) de un conjunto de datos \$( X \$) es:

$$
\mu = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

Donde \$( n \$) es el número de datos. En este caso, \$( n = 12 \$) meses.

Sustituyendo los valores:

$$
\mu = \frac{50 + 48 + 52 + 47 + 55 + 49 + 53 + 51 + 46 + 54 + 50 + 56}{12}
$$

$$
\mu = \frac{611}{12} \approx 50.92
$$

La **media de los ingresos mensuales** es de aproximadamente **50.92 miles de dólares**.

### Paso 2: Cálculo de la desviación estándar

La fórmula para la desviación estándar \$( \sigma \$) de un conjunto de datos es:

$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (X_i - \mu)^2}
$$

Sustituyendo los valores:

$$
\sigma = \sqrt{\frac{1}{12} \left[ (50 - 50.92)^2 + (48 - 50.92)^2 + \dots + (56 - 50.92)^2 \right]}
$$

Realizando los cálculos:

$$
\sigma \approx 3.02
$$

La **desviación estándar** de los ingresos mensuales es de aproximadamente **3.02 miles de dólares**.

### Pregunta 2: ¿Qué mes se desvía más de la media?

Para determinar qué mes se desvía más de la media, calculamos la diferencia absoluta entre los ingresos de cada mes y la media \( \mu \):

- Enero: \$( |50 - 50.92| \approx 0.92 \$)
- Febrero: \$( |48 - 50.92| \approx 2.92 \$)
- Marzo: \$( |52 - 50.92| \approx 1.08 \$)
- Abril: \$( |47 - 50.92| \approx 3.92 \$)
- Mayo: \$( |55 - 50.92| \approx 4.08 \$)
- Junio: \$( |49 - 50.92| \approx 1.92 \$)
- Julio: \$( |53 - 50.92| \approx 2.08 \$)
- Agosto: \$( |51 - 50.92| \approx 0.08 \$)
- Septiembre: \$( |46 - 50.92| \approx 4.92 \$)
- Octubre: \$( |54 - 50.92| \approx 3.08 \$)
- Noviembre: \$( |50 - 50.92| \approx 0.92 \$)
- Diciembre: \$( |56 - 50.92| \approx 5.08 \$)

El **mes de diciembre** se desvía más de la media con una diferencia de **5.08 miles de dólares**.

### Interpretación

El hecho de que diciembre se desvíe más de la media indica un posible aumento significativo en los ingresos, que podría estar relacionado con factores estacionales, como las ventas de fin de año. Este aumento en diciembre podría ser una oportunidad para la empresa de capitalizar sobre eventos especiales o promociones.

## Resumen:

- **Media de los ingresos mensuales**: 50.92 miles de dólares
- **Desviación estándar de los ingresos mensuales**: 3.02 miles de dólares
- **Mes que más se desvía de la media**: Diciembre, con una desviación de 5.08 miles de dólares


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ingresos mensuales (en miles de dólares)
ingresos_mensuales = np.array([50, 48, 52, 47, 55, 49, 53, 51, 46, 54, 50, 56])
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

# Cálculo de la media y desviación estándar
media_ingresos = np.mean(ingresos_mensuales)
desviacion_estandar = np.std(ingresos_mensuales)

# Mostrar resultados
print(f"Media de los ingresos mensuales: {media_ingresos:.2f} miles de dólares")
print(f"Desviación estándar de los ingresos mensuales: {desviacion_estandar:.2f} miles de dólares")

# Calcular la desviación respecto a la media
desviacion = np.abs(ingresos_mensuales - media_ingresos)
mes_mayor_desviacion = meses[np.argmax(desviacion)]
mayor_desviacion = np.max(desviacion)

print(f"El mes que más se desvía de la media es: {mes_mayor_desviacion}, con una desviación de {mayor_desviacion:.2f} miles de dólares")

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.bar(meses, ingresos_mensuales, color='skyblue', label="Ingresos Mensuales")
plt.axhline(y=media_ingresos, color='green', linestyle='--', label=f'Media: {media_ingresos:.2f}')
plt.title('Ingresos Mensuales y Desviación Estándar')
plt.xlabel('Meses')
plt.ylabel('Ingresos (miles de dólares)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
